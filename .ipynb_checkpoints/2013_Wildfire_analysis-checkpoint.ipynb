## Creating dataset for wildfires and stations

In this notebook, we will import three classes of data. First we will import our climate and station data from NOAA. Then we will import our wildfire data and create target data for classification. Finally, we will start to look atwhat questions can be asked of this data.

```python
import noaa_data as noaa
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import requests
```

First we will import in our station data. We will focus on climate data from 2012 and wildfire data from 2013. This is just to help make the data smaller for this test notebook.

```python
#This will give us wildfires that are greater than a 50 acre burn area
wf = noaa.get_wf_data(r'Cal_Wildfires/California_Fire_Incidents.csv', 50)
```

```python
wf_2013 = wf.loc[wf['ArchiveYear']==2013]
```

```python
#Let's plot the data so we can see where the fires are

plt.figure(figsize=(10,12))

sns.scatterplot(wf_2013['Longitude'], wf_2013['Latitude'], size=wf_2013['AcresBurned'], sizes = (50,500), alpha = 0.6 )

plt.xlabel("Longitude", fontsize=16)
plt.ylabel("Latitude", fontsize=16)

plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.distplot(wf_2013['AcresBurned'].values, 
             vertical = True, kde = True)

plt.xlabel('Frequency')
plt.ylabel('Acres Burned')
```

```python
wf_2013.shape
```

It looks like we have a relatively small dataset with a large majority of small fires, and a few very large fires.

Now, we are going to import in our station data and climate data to station data. For this dataframe we will focus on average temperature data.

```python
temp_df = noaa.get_mo_data_noaa('TAVG', 2012, 2012)
```

```python
temp_df.sample(10)
```

The information for this dataset can be found here "https://www.ncdc.noaa.gov/cdo-web/datasets"

Let's add precipitaiton, max temperature and extreme max temperature data as well.

```python
prcp_df = noaa.get_mo_data_noaa('PRCP', 2012, 2012)
tmax_df = noaa.get_mo_data_noaa('TMAX', 2012, 2012)
temax_df = noaa.get_mo_data_noaa('EMXT', 2012, 2012)
```

Now, lets get all our sataion location data

```python
loc_df = noaa.get_stat_noaa(2012, 2012)
```

Now that we have all this, let's combine it into a dataframe

```python
tot_data = temp_df.join(tmax_df, how='outer')
tot_data.head()
```

```python
tot_data = tot_data.join(temax_df, how= 'outer')
tot_data.head()
```

```python
tot_data = tot_data.join(prcp_df, how= 'outer')
```

```python
tot_data.sample(10)
```

```python
loc_df.sample(10)
```

```python
tot_data = tot_data.join(loc_df.set_index('station'), how='outer')
```

```python
tot_data.sample(5)
```

Now we will use <code> noaa.get_close </code> to find stations within 7.0 km of the wildfires in 2013

```python
bad_data = tot_data.loc[tot_data.latitude.isnull()==True].index
```

```python
tot_data = tot_data.drop(bad_data)
```

```python
plt.figure(figsize=(10,12))

sns.scatterplot(wf_2013['Longitude'], wf_2013['Latitude'], size=wf_2013['AcresBurned'], sizes = (50,500) )
sns.scatterplot(tot_data_copy.loc[tot_data_copy['target']==1]['longitude'], tot_data_copy.loc[tot_data_copy['target']==1]['latitude'], sizes=(50))


plt.xlim((-130, -110))
plt.ylim(30,45)
plt.xlabel("Longitude", fontsize=16)
plt.ylabel("Latitude", fontsize=16)

plt.legend({'Wildfires', 'Stations'})




plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.distplot(tot_data_copy.loc[tot_data_copy.target == 1].TAVG.mean(level=1).dropna())
sns.distplot(tot_data_copy.loc[tot_data_copy.target == 0].TAVG.mean(level=1).dropna())

plt.legend({'Wildfire', 'Non-fire'})

plt.xlabel('Mean Temperature')
plt.ylabel('Freq')
```

```python
plt.figure(figsize = (10,5))
sns.distplot(tot_data_copy.loc[tot_data_copy.target == 1].PRCP.mean(level=1).dropna())
sns.distplot(tot_data_copy.loc[tot_data_copy.target == 0].PRCP.mean(level=1).dropna())

plt.legend({'Wildfire', 'Non-fire'})


plt.xlabel('Mean Prcipitation')
plt.ylabel('Freq')
```
