```python
def get_close(wf_frame, stat_frame, cutoff):
    stat_list = np.zeros(stat_frame.coords.shape)
    stat_list = pd.DataFrame(data=stat_list, index= stat_frame.index)
    
    ref_year = int(wf_2019.ArchiveYear.iloc[0])
    
    a_frame = stat_frame.loc[idx[:str(ref_year - 1)+'-12-01',:]]
    b_frame = stat_frame.drop(a_frame.index)
    
    for wf_index, wf_row in wf_frame.iterrows():
        for index, row in a_frame.iterrows():
            if wf_row.Latitude - row.latitude < abs(0.5) and wf_row.Longitude - row.longitude < abs(0.5):
                dist = distance(wf_row.coords, row.coords).km
                if dist < cutoff:
                    stat_list.loc[idx[index]] = int(1)

                    
    for wf_index, wf_row in wf_frame.iterrows():
        for index, row in b_frame.iterrows():
            if wf_row.Started >= pd.to_datetime(index[0]):
                if wf_row.Latitude - row.latitude < abs(0.5) and wf_row.Longitude - row.longitude < abs(0.5):
                    dist = distance(wf_row.coords, row.coords).km
                    if dist < cutoff:
                        stat_list.loc[idx[index]] = int(1)
                        
    return stat_list
```

Ok, now let's do it on a k-fold split.

```python
# import kfold
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
```

```python
kfold = StratifiedKFold(5,shuffle = True,random_state = 440)
```

```python
def get_acc(model,X,y):
    pred = model.predict(X)
    return np.sum(pred == y)/len(y)
```

```python
max_neighbors = 20

accs = np.zeros((5,max_neighbors))

for i in range(1,max_neighbors+1):
    knn = KNeighborsClassifier(i)
    j = 0
    for train_index, test_index in kfold.split(X_train,y_train):
        # Get the cv train test split
        X_train_train, X_train_test = X_train.iloc[train_index], X_train.iloc[test_index]
        y_train_train, y_train_test = y_train.iloc[train_index], y_train.iloc[test_index]
        
        # Cloning the regression makes a fresh regression 
        # model for each run
        clone_knn = clone(knn)
        
        # fit the model
        clone_knn.fit(X_train_train, y_train_train.ravel())
        
        accs[j,i-1] = get_acc(clone_knn, X_train_test, y_train_test.ravel())
        
        j=j+1
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(14,8))

plt.plot(range(1,max_neighbors+1), 100*np.mean(accs, axis=0))

plt.xlabel("Number of Neighbors", fontsize=16)
plt.ylabel("Average CV Accuracy (%)", fontsize=16)

plt.savefig('KNN_fires.png')

plt.show()
```
