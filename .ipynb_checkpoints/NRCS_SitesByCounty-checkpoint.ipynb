```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from NRCS_report import Report
```

```python
#Make the Report object, just load in the previously obtained SiteID info
df_sites = pd.read_csv('SiteIDs.csv')
rep = Report(SiteIDs = df_sites)

rep.SiteIDs.sample(5) #this is the same as saying df_sites.sample(5)
```

```python
#Now lets get all the sites in Inyo county CA
CountyName = 'Inyo'

I_c = rep.SiteIDs['county'] == CountyName
Nsite = I_c.sum()
print('There are',Nsite,'sites in',CountyName,'county')
```

```python
df_county = rep.SiteIDs.loc[I_c]
df_county
```

```python
#Lets try and get a report for the first two sites which are scan sites
IndSelect = [0,8] #station index in above station list df_county to use
Nselect = len(IndSelect)
Sites = [None]*Nselect #here we will write the input for GetReports. Even for 1 site, the format must be a list of 
    #three element lists where each element in the internal lists are strings corresponding to the site ID
    #[State(str),ID #(str),ntwk (str)]
NL=0
for ii in IndSelect:
    Sites[NL] = [None]*3
    Sites[NL][0] = df_county['state'].iloc[ii]
    Sites[NL][1] = df_county['site_id'].iloc[ii]
    Sites[NL][2] = df_county['ntwk'].iloc[ii]
    NL+=1
    
print(Sites)
```

```python
#Now we can get the report. Lets just use the default inputs other than SiteIDs
rep.GetReport(SiteIDs=Sites)
```

```python
#print out results
rep.df_rep #note that if you have sites which all have no data for every entry the dataframe returns blank
    #(as the HTML output table is blank as well)
```
