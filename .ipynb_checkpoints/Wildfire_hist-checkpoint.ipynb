```python
from NRCS_report import Report
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.
wf_data = pd.read_csv('Cal_Wildfires\California_Fire_Incidents.csv')
```

```python
wf_data.columns
```

```python
wf_frame = wf_data[['AcresBurned', 'ArchiveYear', 'Counties', 'Extinguished', 'Latitude', 'Longitude','Name', 'Started']]
```

```python
small_fires = wf_frame.loc[wf_frame['AcresBurned']<50]
```

```python
wf_frame = wf_frame.drop(small_fires.index)
```

```python
wierd = wf_frame.loc[(wf_frame['Latitude']>45) | (wf_frame['Latitude']<30) | (wf_frame['Longitude']>0) ].index
print(wierd.shape)
```

```python
wf_frame = wf_frame.drop(wierd)
```

```python
import seaborn as sns
plt.figure(figsize=(10,12))

sns.scatterplot(wf_frame['Longitude'], wf_frame['Latitude'], size=wf_frame['AcresBurned'], sizes = (5,500), alpha = 0.8 )

plt.xlabel("Longitude", fontsize=16)
plt.ylabel("Latitude", fontsize=16)

plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.distplot(wf_frame.loc[(wf_frame['AcresBurned']>100)& (wf_frame['ArchiveYear']>2018)]['AcresBurned'].values, vertical = True, kde = False)
```
