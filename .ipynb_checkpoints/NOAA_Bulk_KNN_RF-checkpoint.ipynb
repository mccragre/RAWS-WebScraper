```python
def get_close(wf_frame, stat_frame, cutoff):
    stat_list = np.zeros(stat_frame.coords.shape)
    stat_list = pd.DataFrame(data=stat_list, index= stat_frame.index)
                 
    for wf_index, wf_row in wf_frame.iterrows():
        for index, row in stat_frame.iterrows():
            if wf_row.Latitude - row.latitude < abs(1) and wf_row.Longitude - row.longitude < abs(1):
                dist = distance(wf_row.coords, row.coords).km
                if dist < cutoff:
                    stat_list.loc[idx[index]] = int(1)
                        
    return stat_list
```

```python
target_frame = pd.DataFrame()

targets = [5, 7.5, 10, 12.5, 15, 17.5, 20]

for target in targets:
    t = get_close(wf_2019, unique_stations, target)
    target_frame = pd.concat([target_frame, t], axis='columns')


target_frame
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler, QuantileTransformer, Normalizer, RobustScaler
from sklearn.pipeline import Pipeline
```

```python
# import kfold
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
```

```python
kfold = StratifiedKFold(5,shuffle = True,random_state = 440)
```

```python
## Put More Code Here
# Making function to calculate various stats
## Accuracy
from sklearn.metrics import accuracy_score
## Precision
from sklearn.metrics import precision_score
## Recall
from sklearn.metrics import recall_score
```

```python
max_neighbors = 50
kfold = StratifiedKFold(5,shuffle = True,random_state = 440)
accs = np.zeros((5,max_neighbors))
precs = np.zeros((5,max_neighbors))
recs = np.zeros((5,max_neighbors))

for i in range(1,max_neighbors+1):
    knn = KNeighborsClassifier(i, p=3)
    j = 0
    for train_index, test_index in kfold.split(X_train,y_train):
        # Get the cv train test split
        X_train_train, X_train_test = X_train.iloc[train_index], X_train.iloc[test_index]
        y_train_train, y_train_test = y_train.iloc[train_index], y_train.iloc[test_index]
        
        # Cloning the regression makes a fresh regression 
        # model for each run
        clone_knn = clone(knn)
        
        # fit the model
        clone_knn.fit(X_train_train, y_train_train.ravel())
        
        y_predict = clone_knn.predict(X_train_test)
        
        accs[j,i-1] = accuracy_score(y_train_test, y_predict)
        precs[j,i-1] = precision_score(y_train_test,y_predict)
        recs[j,i-1] = recall_score(y_train_test,y_predict)
        
        j=j+1
```

```python
plt.figure(figsize=(14,10))

plt.plot(range(1,max_neighbors+1), np.mean(accs,axis=0),'b-o', label="Accuracy")
plt.plot(range(1,max_neighbors+1), np.mean(precs,axis=0),'r-o', label="Precision")
plt.plot(range(1,max_neighbors+1), np.mean(recs,axis=0),'g-o', label="Recall")

plt.ylabel("Mean CV Performance Score",fontsize=16)
plt.xlabel("Max Depth",fontsize=16)

plt.legend(fontsize=16)

plt.show()
```

```python
max_neighbors = 50
kfold = StratifiedKFold(5,shuffle = True,random_state = 440)
accs = np.zeros((5,max_neighbors))
precs = np.zeros((5,max_neighbors))
recs = np.zeros((5,max_neighbors))

for i in range(1,max_neighbors+1):
    knn = KNeighborsClassifier(i)
    j = 0
    for train_index, test_index in kfold.split(X_train,y_train):
        # Get the cv train test split
        X_train_train, X_train_test = X_train.iloc[train_index], X_train.iloc[test_index]
        y_train_train, y_train_test = y_train.iloc[train_index], y_train.iloc[test_index]
        
        # Cloning the regression makes a fresh regression 
        # model for each run
        clone_knn = clone(knn)
        
        # fit the model
        clone_knn.fit(X_train_train, y_train_train.ravel())
        
        y_predict = clone_knn.predict(X_train_test)
        
        accs[j,i-1] = accuracy_score(y_train_test, y_predict)
        precs[j,i-1] = precision_score(y_train_test,y_predict)
        recs[j,i-1] = recall_score(y_train_test,y_predict)
        
        j=j+1
```

```python
plt.figure(figsize=(14,10))

plt.plot(range(1,max_neighbors+1), np.mean(accs,axis=0),'b-o', label="Accuracy")
plt.plot(range(1,max_neighbors+1), np.mean(precs,axis=0),'r-o', label="Precision")
plt.plot(range(1,max_neighbors+1), np.mean(recs,axis=0),'g-o', label="Recall")

plt.ylabel("Mean CV Performance Score",fontsize=16)
plt.xlabel("Max Depth",fontsize=16)

plt.legend(fontsize=16)

plt.show()
```
